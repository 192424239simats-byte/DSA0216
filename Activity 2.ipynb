```python
# Install OpenCV (Colab usually has it, but this ensures it)
!pip install opencv-python

import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow

# ==============================
# STEP 1: Upload image from your device
# ==============================
uploaded = files.upload()

# Get the uploaded filename
image_path = next(iter(uploaded))

# ==============================
# STEP 2: Read the image
# ==============================
image = cv2.imread(image_path)

# ==============================
# STEP 3: Convert to grayscale
# ==============================
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# ==============================
# STEP 4: Reduce noise
# ==============================
blur = cv2.GaussianBlur(gray, (5, 5), 0)

# ==============================
# STEP 5: Edge detection
# ==============================
edges = cv2.Canny(blur, 50, 150)

# ==============================
# STEP 6: Contour tracing
# ==============================
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# ==============================
# STEP 7: Draw contours
# ==============================
output = image.copy()
cv2.drawContours(output, contours, -1, (0, 255, 0), 2)

# ==============================
# STEP 8: Show results
# ==============================
print("Original Image")
cv2_imshow(image)

print("Edges")
cv2_imshow(edges)

print("Contour Tracing Output")
cv2_imshow(output)

print("Total contours detected:", len(contours))
```
